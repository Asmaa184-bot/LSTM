```python
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def tanh(x):
    return np.tanh(x)
```

```python
X = [1,2,3]

print("Input Sequence:", X)
```

```python
Wf, Whf, bf = 0.5, 0.1, 0
Wi, Whi, bi = 0.6, 0.2, 0
Wc, Whc, bc = 0.7, 0.3, 0
Wo, Who, bo = 0.8, 0.4, 0
```

```python
Wy = 4
by = 0
```

```python
h = 0
C = 0
```

```python
for t,x in enumerate(X):

    # Forget gate
    f = sigmoid(Wf*x + Whf*h + bf)

    # Input gate
    i = sigmoid(Wi*x + Whi*h + bi)

    # Candidate cell state
    C_tilde = tanh(Wc*x + Whc*h + bc)

    # Update cell state
    C = f*C + i*C_tilde

    # Output gate
    o = sigmoid(Wo*x + Who*h + bo)

    # Hidden state
    h = o * tanh(C)

    print("Time step",t+1)
    print("Forget gate f =",round(f,3))
    print("Input gate i =",round(i,3))
    print("Cell state C =",round(C,3))
    print("Hidden state h =",round(h,3))
    print()
```

```python
x = 4

f = sigmoid(Wf*x + Whf*h)
i = sigmoid(Wi*x + Whi*h)
C_tilde = tanh(Wc*x + Whc*h)

C = f*C + i*C_tilde
o = sigmoid(Wo*x + Who*h)
h = o * tanh(C)
```

```python
y = Wy*h + by

print("Predicted value =",y)
```
